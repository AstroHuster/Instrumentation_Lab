```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# log log graph paper
plt.plot((0.1,10),(0.01,1), ',')
plt.grid(visible=True, which='both', axis='both')
plt.xscale('log')
plt.xlim(0.1, 10)
plt.xlabel('Frequency (Hz)')
plt.yscale('log')
plt.ylim(0.01, 1)
plt.ylabel("$|H(f)|^2$")
plt.savefig("Lab06-Filters/log-log_graph_paper.png", dpi=300)
pass
```

```python
# log log graph paper
plt.plot((0.1,10),(0.01,1), ',')
plt.grid(visible=True, which='both', axis='both')
plt.xscale('linear')
plt.xlim(0.1, 10)
plt.xlabel('Frequency (Hz)')
plt.xticks(list(range(0,11)))
plt.yscale('linear')
plt.ylim(0.01, 1)
plt.ylabel("$|H(f)|^2$")
plt.yticks(list(np.arange(0,1.1,.1)))

plt.savefig("Lab06-Filters/linear_graph_paper.png", dpi=300)
pass
```
