```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from glob import glob

%matplotlib inline
```

```python
files = glob("./Data/*.csv")
files.sort()
for (i, f) in enumerate(files):
    print(f"{i:3d} => {f}")
i_file = int(input("Enter file number: "))
data_file = files[i_file]
base_name = data_file[data_file.rfind('/')+1:data_file.rfind('.')]
print("Run ", base_name)
```

```python
(time, phase, x) = np.genfromtxt(data_file, delimiter=',', comments='#').T
with open(data_file, 'r') as fp:
    line = fp.readline().strip()
FREQ = float(line[line.rfind(' '):])
print(FREQ)
```

```python
plt.plot(time, x, '.-')
plt.title(f"Run {base_name}")
plt.xlabel("Time (s)")
plt.ylabel("X (mm)")
plt.grid()
```

# Fit Data to Sine

```python
def SineModel(time, amp, freq, t0):
    return amp * np.sin(2 * np.pi * freq * (time - t0))

amp_start = 0.5 * (np.max(x) - np.min(x))
freq_start = FREQ
t0_start = 0.0
```

```python
paramsFit, paramsCovariance = \
  curve_fit(SineModel, time, x, p0=[amp_start, freq_start, t0_start])
print("paramsFit = ", paramsFit)
print("paramsCovariance =\n", paramsCovariance)
[amp_fit, freq_fit, t0_fit] = paramsFit
```

```python
plt.plot(time, x, '.', label="Data")
plt.plot(time, SineModel(time, amp_start, freq_start, t0_start), '-', label="Guess")
plt.plot(time, SineModel(time, amp_fit, freq_start, t0_start), '-', lw=5, alpha=0.2, label="Best Fit")
plt.title(f"Run {base_name}, Set Freq = {FREQ:.4f}, Fit Freq = {freq_fit:.4f}")
plt.xlabel("Time (s)")
plt.ylabel("X (mm)")
plt.legend()
plt.grid()
plt.savefig("Figures/"+base_name+".png", dpi=150)
```
