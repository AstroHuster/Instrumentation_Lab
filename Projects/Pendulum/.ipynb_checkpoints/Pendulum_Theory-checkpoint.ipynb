# The Physical Pendulum
<img style="float: right; width: 300px" src="physical_pendulum.png">
A rigid object supported by an axis acts like a simple pendulum with a mass, $m$, and 
and effective length 
$$L_{eff} = \frac{I}{m L}$$
where $I$ is the rotational inertia, and $L$ is the distance from the pivot to the 
center of mass.

The mass is easily measured, but the effective length requires 
all parts of a physical pendulum to carefully measure with respect to the axis and the 
rotational inertia, $I$, calculated. I propose to use a steel bar for which $I$ can be 
calculated and a microcontroller/sensor which has a mass of about 10 g as the pendulum. 
This will also be an exercise for the students.

## Equations of Motion 
The torque around the pivot is 
$$\tau = m g L \sin \theta$$
and 
$$
I \frac{d^2 \theta}{d t^2} = m g L \sin \theta
$$

This can be set up as two first order differential equations
$$
\begin{align}
\frac{d \theta}{d t} &= \omega \\
\frac{d \omega}{d t} &= \frac{m g L}{I} \sin \theta
\end{align}
$$

We can define an angular frequency $\omega_0 = \sqrt{\frac{m g L}{I}} = 2 \pi f_0$
giving the final form for the differential equations as 

$$
\begin{align}
\frac{d \theta}{d t} &= \omega \\
\frac{d \omega}{d t} &= \omega_0^2 \sin \theta
\end {align}
$$

The variable $f_0$ is the *natural frequency* of the pendulum at small amplitudes.

# Small Amplitude Solutions

For small amplitudes $\sin \theta \approx \theta$. The equations of motion become 
$$
\begin{align}
\frac{d \theta}{d t} &= \omega \\
\frac{d \omega}{d t} &= \omega_0^2 \theta
\end {align}
$$
which have the general solution 
$$
\theta(t) = \theta_0 \cos(\omega_0 t + \phi)
$$
where $\theta_0$ is the initial angle and $\phi$ is the initial phase. If the initial 
condition is $\theta(0) = \theta_0$ and $\omega_0 = 0$, then $\phi = 0$.

For this inital condition
$$
\begin{align}
\theta(t) &= \theta_0 \cos(\omega_0 t) \\
\omega(t) &= -\theta_0 \omega_0 \sin(\omega_0 t)
\end{align}
$$


# Velocity and Acceleration - Tangential and Radial 

The proposed pendulum will have the sensor at the center of mass. 
The IMU sensor has a gyroscope, which directly measures 
angular velocity, $\omega$ and and accelerometer which measures 
the three-dimensional acceleration. The pendulum is constrained to 
move in a plane, so the relevant accelerations are 
$a_{rad}$ and $a_{tan}$.

For an object in arbitrary circular motion the radial velocity is zero and the tangential 
velocity is 
$$
v_{tan}(t) = L \omega(t)
$$

The tangential acceleration is 
$$
a_{tan}(t) = L \frac{d \omega}{d t}
$$

and the tangential acceleration is
$$
a_{rad}(t) = -L \omega^2(t)
$$

## Simple Harmonic Example

Going back to the previous section, for small amplitudes we can get formulas 
for these. They are

$$
v_{tan}(t) = -\theta_0 \omega_0 L \sin(\omega_0 t)
$$
$$
a_{tan}(t) = -\theta_0 \omega_0^2 L \cos(\omega_0 t)
$$
$$
a_{rad}(t) = -L \theta_0^2 \omega_0^2 \sin^(\omega_0 t)
$$

Later I will use these results to validate my numerical solutions to the 
full non-linear equations.

# Numerically Computing Solutions

The python *scipy* module has very good methods to integrate systems of first 
order differential equations. I will use `scipy.integrate.odeint`. The usage is 
```
out = sp.integrate.odeint(pendulum, (th_i, om_i), t_axis)
```
where `pendulum` is a function that for given $\theta$ and $\omega$ returns  
$\frac{d \theta}{dt}$ and $\frac{d \omega}{dt}$

## Import the Modules

The `python` language depends on external modules for particular applications. 
For the work I do, I almost always use the following modules.

````python
import numpy as np
import matplotlib.pylab as plt
import scipy as sp

%matplotlib inline
````

## Parameters of the Pendulum

Next I have a python cell defining the parameters of the pendulum and other 
variables I need to numerically solve the equations of motion.

Note that you always have to give initial conditions. Here I will use 
$\theta_0 = 5 \ \text{degrees}$, which is a small angle, $\theta_0 << 1$ radian.

````python
G = 9.80 # gravity
L = 0.5 # Effective length of pendulum
OM_0 = np.sqrt(G / L) # small angle angular frequency
T_0 = 2 * np.pi / OM_0
# Initial Conditions

th_i_deg = 10.0
om_i_deg =   0.0

th_i = np.deg2rad(th_i_deg)
om_i = np.deg2rad(om_i_deg)

# Times for the numerical solution
t_i = 0.0 # Start time (s)
t_f = 10.0 # Last time (s)
dt = 0.01 # Time resolution
# Next make an array of the times
t_axis = np.linspace(t_i, t_f, round((t_f - t_i) / dt))
````

````python
# Define pendulum function
# Equations of Motion
def pendulum(y, t):
    th = y[0]
    om = y[1]
    return (om, -OM_0**2 * np.sin(th))
````

## Numerically Solve Equations of Motion and Analytic Solutions

````python
out = sp.integrate.odeint(pendulum, (th_i, om_i), t_axis)
th_calc = out[:,0]
om_calc = out[:,1]

th_model = th_i * np.cos(OM_0 * t_axis)
om_model = -th_i * OM_0 * np.sin(OM_0 * t_axis)
````

## Plot both Numerical and Analytic Solutions

````python
plt.subplot(2,1,1)
plt.plot(t_axis, th_calc, label="Calc")
plt.plot(t_axis, th_model, label='Model')
plt.grid()
plt.legend()
plt.title("$θ_0$ = %.1f Degrees" % (th_i_deg))
plt.ylabel("θ (rad)")

# Get fancy with y axis
secax = plt.gca().secondary_yaxis('right', functions=(np.rad2deg, np.deg2rad))
secax.set_ylabel('angle [°]')
secax.set_ylim(-180, 180)
secax.set_yticks((-180,-90,0,90,180))

plt.subplot(2,1,2)
plt.plot(t_axis, om_calc, label="Calc")
plt.plot(t_axis, om_model, label='Model')
plt.xlabel("Time [s]")
plt.ylabel("ω (rad/s)")
plt.grid()
plt.legend()

pass
plt.yticks(np.array((-1,-.5,0,.5,1))*np.pi)
plt.ylim(-np.pi, np.pi)

# Get fancy with y axis
secax = plt.gca().secondary_yaxis('right', functions=(np.rad2deg, np.deg2rad))
secax.set_ylabel('angle [°]')
secax.set_ylim(-180, 180)
secax.set_yticks((-180,-90,0,90,180))

plt.subplot(2,1,2)
plt.plot(t_axis, om_calc, label="Calc")
plt.plot(t_axis, om_model, label='Model')
plt.xlabel("Time [s]")
plt.ylabel("ω (rad/s)")
plt.grid()
plt.legend()

pass
````

# Measured Accelerations

The main issue before us is, "What will the sensor measure?"

The sensor has a three-axis accelerometer and a three-axis gyroscope, 
It directly measures linear acceleration in the radial and tangental 
directions. It also directly measures the rotational velocity.

The main issue about the IMU sensors is that there is no direct way 
to get the angle of the pendulum. Fortunately there are a few IMU's 
with on board processing that output absolute orientation. I ordered 
one of these units and will assess its usage in this experiment.

````python
# Calculate radial and tangental accelerations
## Tangental
om_dot = np.gradient(om_calc, dt)
a_tan = L * om_dot

## Radial
v = L * om_calc
a_rad = -v**2 / L
````

````python
plt.subplot(2,1,1)
plt.plot(t_axis, a_tan, label="a_tan")
plt.grid()
plt.legend()
plt.title("$θ_0$ = %.1f Degrees" % (th_i_deg))
plt.ylabel("Atan (m/s^2)")

plt.subplot(2,1,2)
plt.plot(t_axis, a_rad, label="a_rad")
plt.xlabel("Time [s]")
plt.ylabel("Arad (m/s^2)")
plt.grid()
plt.legend()

pass
````

# Solutions to the Non-Linear Pendulum
From *An analysis of the large amplitude simple pendulum using Fourier series*, 
Brennen Black and Vetri Vel, Am. J. Phys. 91, 776–782 (2023), 
https://doi.org/10.1119/5.0130943

They find an approximate solution by doing Fourier analysis. They assume a 
solution of the form
$$
θ(θ_0, t) = \sum_{n=0}^\infty{\alpha_n} \cos(n Ω t)
$$
where Ω is the period. The fifth order solution is 
$$
\begin{align}
\alpha_1 &= θ_0 + θ_0^3 / 192 + 17 θ_0^5 / 61440 \\
\alpha_3 &= -θ_0^3 / 192 - θ_0^5 / 3072 \\
\alpha_5 &= θ_0^5 / 20480 \\
T &= T_0 / (1 - θ_0^2 / 16 + θ_0^4 / 3072 - 23 θ_0^6 / 737280
\end{align}
$$
where $T = 2 \pi / Ω$.

The cell below codes this function.

````python
def pend5th(th0, t):
    alf1 = th0 + th0**3 / 192 + 17 * th0**5 / 61440
    alf3 = -th0**3 / 192 - th0**5 / 3072
    alf5 = th0**5 / 20480

    T = T_0 / (1.0 - th0**2 / 16 + th0**4 / 3072 - 23 * th0**6 / 737280)
    OMEGA = 2 * np.pi / T
    
    return alf1 * np.cos(OMEGA * t) + alf3 * np.cos(3 * OMEGA * t) + alf5 * np.cos(5 * OMEGA * t)

def omega5th(th0, t):
    alf1 = th0 + th0**3 / 192 + 17 * th0**5 / 61440
    alf3 = -th0**3 / 192 - th0**5 / 3072
    alf5 = th0**5 / 20480

    T = T_0 / (1.0 - th0**2 / 16 + th0**4 / 3072 - 23 * th0**6 / 737280)
    OMEGA = 2 * np.pi / T
    
    return -alf1 * OMEGA * np.sin(OMEGA * t) + \
            -alf3 * 3 * OMEGA * np.sin(3 * OMEGA * t) + \
            -alf5 * 5 * OMEGA * np.sin(5 * OMEGA * t)
````

# Comparison
At $θ_0 < 120°$ the agreement is excellent. The cell below shows the 
result for 135°, which is also decent.

````python

th_i_deg = 135.0
om_i_deg =   0.0

th_i = np.deg2rad(th_i_deg)
om_i = np.deg2rad(om_i_deg)

out = sp.integrate.odeint(pendulum, (th_i, om_i), t_axis)
th_calc = out[:,0]
om_calc = out[:,1]

th_model = th_i * np.cos(OM_0 * t_axis)
th_model5th = pend5th(th_i, t_axis)
om_model = -th_i * OM_0 * np.sin(OM_0 * t_axis)
om_model5th = omega5th(th_i, t_axis)
plt.subplot(2,1,1)
plt.plot(t_axis, th_calc, label="Calc")
plt.plot(t_axis, th_model, label='Model')
plt.plot(t_axis, th_model5th, label='Model5th', lw=5, alpha=0.35)
plt.title("$θ_0$ = %.1f Degrees" % (th_i_deg))
plt.grid()
plt.legend()
plt.ylabel("θ (rad)")
# plt.yticks(np.array((-1,-.5,0,.5,1))*np.pi)
# plt.ylim(-np.pi, np.pi)

# Get fancy with y axis
secax = plt.gca().secondary_yaxis('right', functions=(np.rad2deg, np.deg2rad))
secax.set_ylabel('angle [°]')
secax.set_ylim(-180, 180)
secax.set_yticks((-180,-90,0,90,180))

plt.subplot(2,1,2)
plt.plot(t_axis, om_calc, label="Calc")
plt.plot(t_axis, om_model, label='Model')
plt.plot(t_axis, om_model5th, label='Model5th',lw=5, alpha=0.35)
plt.xlabel("Time [s]")
plt.ylabel("ω (rad/s)")
plt.grid()
plt.legend()

pass
````
