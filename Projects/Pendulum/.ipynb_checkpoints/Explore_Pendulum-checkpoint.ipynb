```python
import numpy as np
import matplotlib.pylab as plt
import scipy as sp

%matplotlib inline
```

```python
# Pendulum Parameters
# Assume rigid, massless rod

G = 9.80
L = 0.5
M = 0.1
I = M * L**2
OM_0 = np.sqrt(G / L)
```

```python
# Initial Conditions

th_i_deg = 15.0
om_i_deg =   0.0

th_i = np.deg2rad(th_i_deg)
om_i = np.deg2rad(om_i_deg)

t_i = 0.0
t_f = 10.0
dt = 0.01
t_axis = np.linspace(t_i, t_f, round((t_f - t_i) / dt))

# Equations of Motion
alpha = G / L
def pendulum(y, t):
    th = y[0]
    om = y[1]
    return (om, -alpha * np.sin(th))

def deg2rad(x):
    return x * np.pi / 180

def rad2deg(x):
    return x * 180 / np.pi
```

```python
out = sp.integrate.odeint(pendulum, (th_i, om_i), t_axis)
th_calc = out[:,0]
om_calc = out[:,1]

th_model = th_i * np.cos(OM_0 * t_axis)
om_model = -th_i * OM_0 * np.sin(OM_0 * t_axis)
```

```python
plt.subplot(2,1,1)
plt.plot(t_axis, th_calc, label="Calc")
plt.plot(t_axis, th_model, label='Model')
plt.grid()
plt.legend()
plt.ylabel(" (rad)")
plt.yticks(np.array((-1,-.5,0,.5,1))*np.pi)
plt.ylim(-np.pi, np.pi)

# Get fancy with y axis
secax = plt.gca().secondary_yaxis('right', functions=(rad2deg, deg2rad))
secax.set_ylabel('angle [deg]')
secax.set_ylim(-180, 180)
secax.set_yticks((-180,-90,0,90,180))

plt.subplot(2,1,2)
plt.plot(t_axis, om_calc, label="Calc")
plt.plot(t_axis, om_model, label='Model')
plt.xlabel("Time [s]")
plt.ylabel("")
plt.grid()
plt.legend()

pass
```

```python
# Calculate radial and tangental accelerations
## Tangental
om_dot = np.gradient(om_calc, dt)
a_tan = L * om_dot

## Radial
v = L * om_calc
a_rad = -v**2 / L
```

```python
# Tangential Acceleration
plt.subplot(2,1,1)
plt.plot(t_axis, a_tan)
plt.ylabel("$a_{tan}$")
plt.title(F"Measured Accelerations (- g), theta_0 = {np.rad2deg(th_i):.1f} deg")
plt.grid()

# Radia Acceleration
plt.subplot(2,1,2)
plt.plot(t_axis, a_rad)
plt.xlabel("Time [s]")
plt.ylabel("$a_{rad}$")
plt.grid()

pass
```

# Add Gravity
The accelerometers measure the kinematic accelerations plus 
the "acceleration of gravity" $G$, or gravitational field strength. 
It is alway magnitude $G$ and in the $-\hat y$ direction.

```python
a_rad_total = a_rad + G * np.cos(th_calc)
a_tan_total = a_tan - G * np.sin(th_calc)
```

```python
# Tangential Acceleration
plt.subplot(2,1,1)
plt.plot(t_axis, a_tan_total)
plt.ylabel("$a_{tan}$")
plt.title(F"Measured Accelerations (- g), theta_0 = {np.rad2deg(th_i):.1f} deg")
plt.grid()

# Radia Acceleration
plt.subplot(2,1,2)
plt.plot(t_axis, a_rad_total)
plt.xlabel("Time [s]")
plt.ylabel("$a_{rad}$")
plt.grid()

pass
```

```python
plt.plot(t_axis, a_rad, label=f"Radial")
plt.plot(t_axis, a_rad_total, label="Radial + Gr")
plt.title("Radial Accerations")
plt.grid()
```

```python
mult = 1.0
plt.plot(t_axis, mult * a_tan, label=f"{mult}*a_tan")
plt.plot(t_axis, a_tan_total, label = "a_tan + G")
plt.title("Tangential Accelerations")
plt.grid()
plt.legend()

pass
```
