```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.linspace(-np.pi/2, np.pi/2, 500)
cos = np.cos(x_axis)
a = (np.pi / 2)**2
cos_appx = -(x_axis**2 - a)/a

plt.plot(x_axis, cos, label="Cos")
plt.plot(x_axis, cos_appx, label="Cos_appx")
plt.legend()
plt.grid()

pass
```

```python
x2_axis = np.linspace(0, 6 * np.pi, 500)

def cos_appx2(x):
    ret = np.zeros_like(x)
    xt = x % (2 * np.pi)
    a = np.pi**2 / 4
    iw = np.where(xt <= 0.5 * np.pi)
    ret[iw] = -(xt[iw]**2 - a)/a
    iw = np.where(xt > 1.5 * np.pi)
    ret[iw] = -((xt[iw] - 2 * np.pi)**2 - a)/a
    iw = np.where((xt > 0.5 * np.pi) & (xt <= 1.5 * np.pi))
    ret[iw] = ((xt[iw] - np.pi)**2 - a)/a
    return ret
    
plt.plot(x2_axis, np.cos(x2_axis), label="Cos")
plt.plot(x2_axis, cos_appx2(x2_axis), label="Cos_appx")
plt.legend()
plt.grid()

pass
```
